```python
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib notebook
```

```python
algos = ['CE', 'Soft Target', 'PPO', 'SAC']
```

```python
def mean_std_from_file(file_name):
    files = os.listdir(file_name)
    txt_patern = "*.txt"
    files = [f for f in files if '.txt' in f]
    files = [f for f in files if not '_0' in f]
    files = [f for f in files if not '_prob' in f]
    
    file0 = files[0]
    data_ = np.loadtxt(file_name+'/'+file0)
    e = data_[:,0]
    
    res = np.ones((e.shape[0],len(files)))
    res[:,0] = data_[:,1]
    res_episodes = e.copy()
    print(files)
    for i in range(1,len(files)):
        data_ = np.loadtxt(file_name+'/'+files[i])
        e = data_[:,0]
        if not np.all(res_episodes == e):
            raise('episode numbers do not match')
        else:
            res[:,i] = data_[:,1]
    
    return e, np.mean(res,axis=1),  np.std(res,axis=1)
```

```python
ece, mce, stdce = mean_std_from_file(algos[0])
est, mst, stdst = mean_std_from_file(algos[1])
eppo, mppo, stdppo = mean_std_from_file(algos[2])
esac, msac, stdsac = mean_std_from_file(algos[3])
```

```python
plt.plot(ece,mce,color='b',label='Deep Cross Entropy')
plt.fill_between(ece,mce-stdce,mce+stdce, color='b', alpha=0.2)
plt.plot(est,mst,color='g',label='Soft Target DQN')
plt.fill_between(est,mst-stdst,mst+stdst, color='g', alpha=0.2)
plt.plot(eppo,mppo,color='r',label='PPO')
plt.fill_between(eppo,mppo-stdppo,mppo+stdppo, color='r', alpha=0.2)
plt.plot(esac,msac,color='gold',label='SAC')
plt.fill_between(esac,msac-stdsac,msac+stdsac, color='gold', alpha=0.2)
plt.ylim([-500,350])



plt.grid()
plt.legend()
plt.xlabel('# trajectories')
plt.ylabel('reward')
```

```python
plt.savefig('comp.png',dpi=300, facecolor='w', bbox_inches ='tight')
```

```python
def window(array):
    return np.convolve(array,np.ones(20)/20,mode='valid')

def mean_std_from_file(file_name):
    files = os.listdir(file_name)
    txt_patern = "*.txt"
    files = [f for f in files if '.txt' in f]
    files = [f for f in files if '_0' in f]
    files = [f for f in files if not '_prob' in f]
    
    file0 = files[0]
    data_ = np.loadtxt(file_name+'/'+file0)
    e = data_[:,0]
    
    res = np.ones((e.shape[0],len(files)))
    res[:,0] = data_[:,1]
    res_episodes = e.copy()
    print(files)
    for i in range(1,len(files)):
        data_ = np.loadtxt(file_name+'/'+files[i])
        e = data_[:,0]
        if not np.all(res_episodes == e):
            raise('episode numbers do not match')
        else:
            res[:,i] = data_[:,1]
    
    return e, np.mean(res,axis=1),  np.std(res,axis=1)

est_, mst_, stdst_ = mean_std_from_file(algos[1])
```

```python
plt.plot(window(est),window(mst),color='g',label='Soft Target DQN')
plt.fill_between(window(est),window(mst-stdst),window(mst+stdst), color='g', alpha=0.2)
plt.plot(window(est_),window(mst_),color='k',label='Soft Target DQN with 0 initial target')
plt.fill_between(window(est_),window(mst_-stdst_),window(mst_+stdst_), color='k', alpha=0.2)
plt.plot(window(eppo),window(mppo),color='r',label='PPO')
plt.fill_between(window(eppo),window(mppo-stdppo),window(mppo+stdppo), color='r', alpha=0.2)
plt.plot(window(esac),window(msac),color='gold',label='SAC')
plt.fill_between(window(esac),window(msac-stdsac),window(msac+stdsac), color='gold', alpha=0.2)
plt.ylim([-500,350])



plt.grid()
plt.legend()
plt.xlabel('# trajectories')
plt.ylabel('reward')
```
