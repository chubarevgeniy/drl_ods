```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
files = os.listdir()
```

```python
files = ['Laplace_pol_q_0.6__tr_n_600__smoth_0.16.txt',
 'Laplace_pol_q_0.8__tr_n_600__smoth_0.04.txt',
 'Laplace_pol_q_0.8__tr_n_600__smoth_0.08.txt',
 'Laplace_pol_q_0.8__tr_n_600__smoth_0.16.txt',
 'Policy_pol_q_0.6__tr_n_600__smoth_0.5.txt',
 'Policy_pol_q_0.8__tr_n_600__smoth_0.25.txt',
 'Policy_pol_q_0.8__tr_n_600__smoth_0.5.txt',
 'Policy_pol_q_0.8__tr_n_600__smoth_0.6.txt',
 'Policy_pol_q_0.8__tr_n_600__smoth_0.8.txt',
 'rewrite_pol_q_0.6__tr_n_600.txt',
 'rewrite_pol_q_0.8__tr_n_600.txt',
 'stoc_pol_q_0.6__sample_n_400__traj_per_samp_20.txt']
```

```python
history = {}
for name in files:
    data = np.loadtxt(name)
    history[name] = data.copy()
```

```python
plt.plot(history['rewrite_pol_q_0.6__tr_n_600.txt'],label='rewrite_pol_q_0.6__tr_n_600')
plt.plot(history['rewrite_pol_q_0.8__tr_n_600.txt'],label='rewrite_pol_q_0.8__tr_n_600')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('avg reward per epoch')
plt.grid(True)
plt.savefig('rewrite_pol.png',bbox_inches='tight',dpi=300)
```
